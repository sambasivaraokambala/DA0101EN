<h2>Read Data</h2>

We use pandas.read_csv() function to read the csv file.In the bracket we put the file path along with quotation mark,so that pandas will read the file into
dataframe from that address.The file path can be either an URL or local file address.

Because, the data does not contain headers, we can add an argument headers = None inside the read_csv() function,so that pandas will not automatically
set the first row as header.


you can also assign a dataframe to any variable you create.

```python
# import pandas library
import pandas as pd
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(url,header = None)
```

After reading the dataset, we can use the dataframe.head(n)method to check the top n rows of the dataframe,Where n is an integer.Contrary to the dataframe.head(n),
dataframe.tail(n) will show you the bottom n rows of the dataframe

```python
# Show the first 5 rows of the dataframe using dataframe.head() method

print("The first five rows of the dataframe:")
df.head(5)
```

```python
# show the bottom 10 rows of the dataframe
print("The bottom 10 rows:")
df.tail(10)
```

<h2>Add Headers</h2>

Take a look at our dataset.pandas automatically set the header from integer 0.

To better describe our data we can introduce a header, this information is available at:  <a href="https://archive.ics.uci.edu/ml/datasets/Automobile" target="_blank">https://archive.ics.uci.edu/ml/datasets/Automobile</a>
</p>

<p>
Thus we have to add headers manually
</p>
<p>
Firstly, we create a list "headers" that contain all column names in order
    Then, we use <code>dataframe.columns =headers </code> to replace the headers by the list we created.
    </p>

```python
# Create a headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base",
          "length","width","height","curb-weight","engine-type","num-of-cylinders","enzine-size","fuel-system","bore","stroke","compression-ratio",
          "horse-power","peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n",headers)
```

We replace headers and recheck our data.

```python
df.columns = headers
df.head(20)
```

<p>
Now, we need to replace the '?' symbol with NaN,so that <code>dropna()</code> can remove the missing values.

```python
df1= df.replace('?',np.NaN)
```

<p> We candraop the missing values along the column "Price" as follows</p>

```python
df = df1.dropna(subset=["price","normalized-losses"],axis = 0)
df.head(20)
```

Now, we have sucessfully read the raw dataset and correct headers into dataset.

```python
# printing column names
print(df.columns)
```
